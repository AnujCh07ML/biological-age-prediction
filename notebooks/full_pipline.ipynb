```python
import pandas as pd
import seaborn as sns
from pathlib import Path
```

```python
base_path = Path('/Users/anuj/Documents/CodeBase/Bioinformatics_projects/biological-age-prediction/data/raw')
```

```python
data = {}
for years in base_path.iterdir():
    year = years.name
    
    for features in years.glob('*.xpt'):
        
        feature_name = features.stem.split("_")
        if feature_name[0] == 'P':
            feature_name = feature_name[1]
        else:
            feature_name = feature_name[0]
        
                           
        df = pd.read_sas(features, format='xport', encoding="latin1")
        
        df['year'] = year
        data[(year, feature_name)] = df
        
```

```python
for key, value in data.items():
    print(f'key = {key} & value = {value.shape}')
```

## Combine yearly data
From each year, extract one feature eg. DEMO and add rest of the features on DSQN.

```python
year = []

for years in data.keys():
    year.append(years[0])

all_years = list(set(year))
all_years
```

### Get features list

```python
features = []

for key in data.keys():
    features.append(key[1])
    
All_features = list(set(features))
All_features
```

### Merge year

```python
merged_data = {}
for year in all_years:
    df_merged = data[(year,'DEMO')].copy()

    for feature in All_features:

        if feature == 'DEMO':
            continue

        if (year, feature) not in data:
            continue

        df_feature = data[(year, feature)]

        df_feature = df_feature.drop(columns=['year'], errors='ignore')

        df_merged = pd.merge(df_merged, df_feature, on = "SEQN", how = 'left')

        merged_data[year] = df_merged
        

df_merged
```

```python
print(merged_data['2015-16'].shape)
print(df_merged.shape)
print(merged_data.keys())
print(df_merged.columns[:20])
```

```python
dfs = []

for year, df_year in merged_data.items():
    df_year['year'] = year
    dfs.append(df_year)


df = pd.concat(dfs, ignore_index=True)
```
